```python
def add_and_check_even_or_odd(number1, number2):
    # Calculate the sum of the two numbers
    result = number1 + number2

    # Check if the result is even or odd
    if result % 2 == 0:
        print(f"The sum {result} is even.")
    else:
        print(f"The sum {result} is odd.")

# Example usage:
num1 = 5
num2 = 7
add_and_check_even_or_odd(num1, num2)
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year_to_check = 2024
if is_leap_year(year_to_check):
    print(f"{year_to_check} is a leap year.")
else:
    print(f"{year_to_check} is not a leap year.")
```

```python
import pandas as pd
df = pd.read_csv('senator_tweets.csv')
```

```python
df.head()
```

```python
# Add a new column 'CharCount' that counts the number of characters in each tweet
df['CharCount'] = df['text'].apply(len)

# Display the updated dataset
print(df)
```

```python
# Calculate the correlation coefficients
correlation = df[['CharCount', 'replies', 'retweets', 'favorites']].corr()

# Display the correlation matrix
print(correlation)
```

```python
# Sort the dataset by 'CharCount' in ascending order
sorted_df = df.sort_values(by='CharCount')

# Get the shortest tweet (first row after sorting)
shortest_tweet = sorted_df.iloc[0]['text']

# Display the shortest tweet
print("The shortest tweet is:", shortest_tweet)
```

```python
# Filter tweets that include the word "immigration"
immigration_tweets = df[df['text'].str.contains('immigration', case=False)]

# Count the number of immigration tweets
num_immigration_tweets = immigration_tweets.shape[0]

# Display the filtered tweets and the count
print("Tweets that include the word 'immigration':")
print(immigration_tweets)
print("\nNumber of immigration tweets:", num_immigration_tweets)
```

```python
# Filter tweets that include the word "Canada"
canada_tweets = df[df['text'].str.contains('Canada', case=False)]

# Count the number of Canada tweets
num_canada_tweets = canada_tweets.shape[0]

# Display the filtered tweets and the count
print("Tweets that include the word 'Canada':")
print(canada_tweets)
print("\nNumber of Canada tweets:", num_canada_tweets)
```

```python
# Convert the 'Date' column to datetime format
df['created_at'] = pd.to_datetime(df['created_at'])

# Filter tweets sent in 2010
tweets_2010 = df[df['created_at'].dt.year == 2010]

# Count the number of tweets sent in 2010
num_tweets_2010 = tweets_2010.shape[0]

# Display the filtered tweets and the count
print("Tweets sent in 2010:")
print(tweets_2010)
print("\nNumber of tweets sent in 2010:", num_tweets_2010)
```

```python
# Convert the 'Date' column to datetime format
df['created_at'] = pd.to_datetime(df['created_at'])

# Extract the month and add it as a new column 'MonthSent'
df['MonthSent'] = df['created_at'].dt.month

# Group by 'MonthSent' and count the number of tweets in each month
monthly_tweet_counts = df.groupby('MonthSent').size().reset_index(name='TweetCount')

# Find the month with the most tweets
most_tweets_month = monthly_tweet_counts.loc[monthly_tweet_counts['TweetCount'].idxmax()]

# Display the monthly tweet counts and the month with the most tweets
print("Monthly Tweet Counts:")
print(monthly_tweet_counts)

print("\nThe month with the most tweets is:")
print(most_tweets_month)
```

```python
# Use the str.split() method to split the 'Tweet' column into words and count them
df['WordCount'] = df['text'].str.split().apply(len)

# Display the updated dataset
print(df)
```

```python
# Calculate the word count for each tweet and add it as a new column 'WordCount'
df['WordCount'] = df['text'].str.split().apply(len)

# Group the dataset by 'State' and calculate the average word count for tweets in each state
average_word_count_by_state = df.groupby('state')['WordCount'].mean().reset_index()

# Find the state with the highest average word count
state_with_longest_tweets = average_word_count_by_state.loc[average_word_count_by_state['WordCount'].idxmax()]

# Display the average word count by state and the state with the longest tweets
print("Average Word Count by State:")
print(average_word_count_by_state)

print("\nThe state with the longest tweets on average is:")
print(state_with_longest_tweets)
```
